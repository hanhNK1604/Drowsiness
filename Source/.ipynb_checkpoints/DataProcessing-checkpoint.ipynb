```python
import os
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision
from PIL import Image
```

```python
close_path = '/content/drive/MyDrive/DEEP_LEARNING/DeepLearning/Project/Drowsiness/RawData/Closed_Eyes/'
close_file = os.listdir(close_path)
print(close_file)
```

```python
#check data
image = Image.open(close_path + close_file[0])
plt.imshow(image)
```

```python
#image shape
print(image.size)
```

```python
print(image)
```

```python
#transform to tensor and check shape of tensor
transform = torchvision.transforms.Compose([
      torchvision.transforms.Resize((128, 128)),
      torchvision.transforms.ToTensor()
])

new_image = transform(image)
print(new_image.shape)
plt.imshow(new_image.view(128, -1))
```

```python
close_image = []
close_label = []
for dir in close_file:
  # print(dir)
  image = Image.open(close_path + dir)
  image = transform(image)
  close_image.append(image)
  close_label.append(1)

print(len(close_image))
```

```python
open_path = '/content/drive/MyDrive/DEEP_LEARNING/DeepLearning/Project/Drowsiness/RawData/Open_Eyes/'
open_file = os.listdir(open_path)
open_image = []
open_label = []
for dir in open_file:
  # print(dir)
  image = Image.open(open_path + dir)
  image = transform(image)
  open_image.append(image)
  open_label.append(0)

print(len(open_image))
```
